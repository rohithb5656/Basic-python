3.What are the main differences between lists and tuples in Python3

1.Mutability: Lists are mutable (can be changed); tuples are immutable (cannot be changed).

2.Syntax: Lists use []; tuples use ().

3.Methods: Lists have more methods for modification; tuples have fewer methods.

4.Performance: Lists are slower; tuples are faster.

5.Memory: Lists use more memory; tuples are memory efficient.

6.Use Case: Use lists when data changes; use tuples for constant data.

7.Hashability: Lists are not hashable; tuples are hashable (if they contain immutable items).

4.Describe how dictionaries store data

Dictionaries in Python store data using a key-value pair system, where each key is unique, and it maps to a corresponding value.

1.Key-Value Pairs: A dictionary is composed of keys and values. The key is like an identifier or label, and the value is the associated data.

Example: {'name': 'Alice', 'age': 30}. Here, 'name' and 'age' are keys, and 'Alice' and 30 are the corresponding values.

2.Hashing: Python dictionaries use a technique called hashing to store keys efficiently. When a key is added to the dictionary, Python computes a hash value for the key. This hash value determines where the key-value pair is stored in memory.

The hash function ensures that lookups are fast, even for large datasets.

3.Uniqueness of Keys: Dictionary keys must be unique. If you try to add a key that already exists, the new value will overwrite the previous one associated with that key.

Example: If you update my_dict = {'name': 'Alice'} with my_dict['name'] = 'Bob', the dictionary becomes {'name': 'Bob'}.

4.Unordered: Dictionaries are unordered collections. The items inside a dictionary do not have a fixed order, meaning that the order of the keys may not be the same when iterating through the dictionary,

5.Why might you use a set instead of a list in Python

1.Uniqueness: A set automatically ensures that all elements are unique, meaning no duplicates are allowed. If you need to store only unique items, a set is ideal.

2.Faster Lookups: Sets are optimized for fast membership testing (checking if an item exists). They can check for membership much faster than lists, especially as the collection grows.

3.Set Operations: If you need to perform operations like unions, intersections, or differences between collections, sets are designed for this, and these operations are faster with sets.

4.Unordered: If you don’t care about the order of elements and just need to store unique items, a set is a better choice because it does not maintain any order.

6.What is a string in Python, and how is it different from a list

A string in Python is a sequence of characters enclosed within single quotes (' ') or double quotes (" "). Strings are used to represent text-based data.
Strings are immutable in Python, meaning once a string is created, its contents cannot be changed. If you modify a string, Python creates a new string instead of modifying the existing one.

1.Data Type:

String: A string is a sequence of characters (text data).

List: A list is an ordered collection of items, which can be of any data type (e.g., integers, strings, other lists).

2.Mutability:

String: Strings are immutable, meaning you cannot change individual characters once the string is created. To change the string, you must create a new one.

List: Lists are mutable, meaning you can modify, add, or remove elements after the list is created.

3.Elements:

String: A string contains only characters (letters, digits, symbols).

List: A list can contain elements of different data types, such as numbers, strings, other lists, and more.

4.Operations:

String: You can perform string-specific operations like concatenation, slicing, and formatting. But you cannot change individual characters directly.

List: Lists allow you to modify, add, or remove elements using methods like append(), remove(), or by direct indexing.

5.Indexing:

Both strings and lists are indexed by position, starting from 0, and you can access individual elements using their indices. However, since strings are immutable, attempting to assign a value to a string index will result in an error, while lists allow such changes.

8.What is a hash table, and how does it relate to dictionaries in Python

A hash table is a data structure that stores data in key-value pairs, where a hash function is used to compute an index (or hash) for each key. This allows for fast access, insertion, and deletion of items.

In Python, dictionaries use hash tables internally. The keys in a dictionary are hashed to quickly find their corresponding values. This is why looking up a value in a dictionary by its key is usually very fast.

Key Points:
Hash Table: Stores key-value pairs and uses a hash function for fast lookups.

Dictionary in Python: A Python dictionary is implemented using a hash table, making access to values by their keys very efficient.

14.How does the “in” keyword work differently for lists and dictionaries.?

In Python, the in keyword behaves differently for lists and dictionaries:

For lists:
in checks if a value exists in the list.
fruits = ['apple', 'banana']
print('apple' in fruits)  # True

For dictionaries:
in checks if a key exists in the dictionary (not the value).
person = {'name': 'Alice', 'age': 25}
print('name' in person)  # True
print('Alice' in person)  # False

.in with lists → checks for values
.in with dictionaries → checks for keys

20.Explain the difference between a list and a dictionary in terms of data retrieval.

The main difference between a list and a dictionary in terms of data retrieval is:

1.List: Retrieves elements by their index (position in the list), which is a numerical value. Access is sequential, and searching for an element by value is slower (O(n) time complexity).

2.Dictionary: Retrieves values by their key, allowing for faster lookups (O(1) time complexity) because keys are hashed. You don't need to know the position; just the key is enough.
